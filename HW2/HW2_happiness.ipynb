```python
happiness[['Happiness Score', 'Standard Error','Economy (GDP per Capita)','Family','Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual', 'year']].corr()
```

This measures the correlation among the numerical variables

```python
pl.imshow(happiness[['Happiness Score', 'Standard Error','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual', 'year']].corr())

pl.xticks([0 , 2, 4, 6, 8], ['Happiness Score', 'Standard Error','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual', 'year'], rotation = 75)
pl.yticks([0, 2, 4, 6, 8,], ['Happiness Score', 'Standard Error','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom', 'Trust (Government Corruption)', 'Generosity','Dystopia Residual', 'year'], rotation = 0 )
pl.colorbar();
```

This is the plot of the correlation of variables. The yellow color shows correlation, dark green from the 0.0 point to lighter green at the 0.8 point show positive values, and dark blue below the 0.0 point show negative values
